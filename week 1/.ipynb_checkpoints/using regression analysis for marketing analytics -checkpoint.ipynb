```python
d['Engaged'] = d['Response'].apply(lambda x: 0 if x == 'No' else 1)
```

```python
d.head()
```

```python
engagement_rate_df = pd.DataFrame(
d.groupby('Engaged').count()['Response'] / d.shape[0] * 100.0
)
```

```python
engagement_rate_df
```

```python
engagement_rate_df.T
```

```python
engagement_by_sales_channel_df = pd.pivot_table(
d, values='Response', index='Sales Channel', columns='Engaged', aggfunc=len
).fillna(0.0)
engagement_by_sales_channel_df.columns = ['Not Engaged', 'Engaged']
```

```python
engagement_by_sales_channel_df
```

```python
engagement_by_sales_channel_df.plot(
kind='pie',
figsize=(15, 7),
startangle=90,
subplots=True,
autopct=lambda x: '%0.1f%%' % x
)
plt.show()
```

```python
ax = d[['Engaged', 'Total Claim Amount']].boxplot(
by='Engaged',
showfliers=False,
figsize=(7,5)
)
ax.set_xlabel('Engaged')
ax.set_ylabel('Total Claim Amount')
ax.set_title('Total Claim Amount Distributions by Engagements')
plt.suptitle("")
plt.show()
```

```python
ax = d[['Engaged', 'Total Claim Amount']].boxplot(
by='Engaged',
showfliers=True,
figsize=(7,5)
)
ax.set_xlabel('Engaged')
ax.set_ylabel('Total Claim Amount')
ax.set_title('Total Claim Amount Distributions by Engagements')
plt.suptitle("")
plt.show()
```

```python
continuous_vars = [
'Customer Lifetime Value', 'Income', 'Monthly Premium Auto',
'Months Since Last Claim', 'Months Since Policy Inception',
'Number of Open Complaints', 'Number of Policies',
'Total Claim Amount'
]
```

```python
import statsmodels.api as sm
x=d['Engaged']
y=d[continuous_vars]
```

```python
logit = sm.Logit(x,y)
```

```python
logit_fit=logit.fit()
```

```python
logit_fit.summary()
```

```python
gender_values,gender_labels=d['Gender'].factorize()
```

```python
gender_values
```

```python
gender_labels
```

```python
categories = pd.Categorical(
d['Education'],
categories=['High School or Below', 'Bachelor', 'College', 'Master', 'Doctor']
)
```

```python
d['GenderFactorized'] = gender_values
d['EducationFactorized'] = categories.codes
```

```python
logit = sm.Logit(
d['Engaged'],
d[[
'GenderFactorized',
'EducationFactorized'
]]
)
logit_fit = logit.fit()
```

```python
logit_fit.summary()
```

```python
logit = sm.Logit(
d['Engaged'],
d[['Customer Lifetime Value',
'Income',
'Monthly Premium Auto',
'Months Since Last Claim',
'Months Since Policy Inception',
'Number of Open Complaints',
'Number of Policies',
'Total Claim Amount',
'GenderFactorized',
'EducationFactorized'
]]
)
logit_fit = logit.fit()
```

```python
logit_fit.summary()
```
